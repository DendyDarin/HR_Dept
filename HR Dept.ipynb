**Import Libraries and Data Set**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('Human_Resources.csv')
```

```python
data
```

```python
data.info()
```

```python
data.describe()
```

```python
#In Attrition columns we find Yes/No values, we can change it to integer 1/0
#use lambda function

data['Attrition'] = data['Attrition'].apply(lambda x:1 if x == 'Yes' else 0)
```

```python
data
```

```python
#Apply it for Over18 and OverTime columns

data['Over18'] = data['Over18'].apply(lambda x:1 if x == 'Y' else 0)
data['OverTime'] = data['OverTime'].apply(lambda x:1 if x == 'Yes' else 0)
```

```python
data.head()
```

```python
data.describe()
```

```python
# Checking missing data use seaborn

sns.heatmap(data.isnull(), yticklabels=False, cbar=False, cmap='Blues')
```

**Visualizing Dataset**

```python
# Visualization using histogram

data.hist(bins=30, figsize=(20,20), color='r')
```

```python
# Drop unnecessary columns that have fixed value, like 'EmployeeCount', 'StandarHours' and 'Over18'
# inplace = True means the dropped columns has gone from memory and won't count in the future
data.drop(['EmployeeCount', 'StandardHours', 'Over18', 'EmployeeNumber'], axis=1, inplace=True)
```
